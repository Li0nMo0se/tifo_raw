```python
from skimage import io
from skimage import exposure
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Grayscale histogram

```python
filename = "../data/image.result.pgm"

img = io.imread(filename)
img.shape
```

```python
hist, bins = exposure.histogram(img)
hist.shape
```

```python
plt.bar(bins, hist)
plt.show()
```

# RGB histogram

```python
def compute_plot_rgb_hist(in_img):
    red_hist, red_bins  = exposure.histogram(in_img[...,0])
    green_hist, green_bins  = exposure.histogram(in_img[...,1])
    blue_hist, blue_bins  = exposure.histogram(in_img[...,2])
    
    plt.figure(figsize=(15,3))
    plt.subplot(131)
    plt.bar(red_bins, red_hist, color='red')
    plt.subplot(132)
    plt.bar(green_bins, green_hist, color='green')
    plt.subplot(133)
    plt.bar(blue_bins, blue_hist, color='blue')
    plt.show()
```

## From interpolation

```python
filename = "../data/image.result.ppm"

img = io.imread(filename)
img.shape
```

```python
compute_plot_rgb_hist(img)
```

## With white balancing

```python
filename = "../data/image.whitebalance.result.ppm"

img = io.imread(filename)
img.shape
```

```python
compute_plot_rgb_hist(img)
```

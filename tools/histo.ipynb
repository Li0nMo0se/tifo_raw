```python
from skimage import io
from skimage import exposure
import matplotlib.pyplot as plt

%matplotlib inline
```

```python

filename = "image.result.pgm"

img = io.imread(filename)
img.shape
```

```python
hist, bins = exposure.histogram(img)
hist.shape, bins.shape
```

```python
plt.plot(bins, hist)
plt.show()
```
